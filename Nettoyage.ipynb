```python
from os import listdir
```

```python
# Affichage des noms des fichiers textes
noms_fichiers = listdir("Speeches")
noms_fichiers
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(index=range(len(noms_fichiers)), columns=['Nom_du_Fichier','Contenu'])
df.head()
```

```python
for i,f_name in enumerate(noms_fichiers):
    f = open("Speeches/" + f_name, "r", encoding="utf-8")
    f_content = f.readlines()
    f.close()
    
    df.at[i, "Nom_du_Fichier"] = f_name
    df.at[i, "Contenu"] = f_content[0]
```

```python
# Liste des mots possibles
mois = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
mois
```

```python
def SeparerVille(v):
    for m in mois:
        if (m in v):
            ville = v[:v.find(m)]
            return ville
```

```python
# Application de la fonction SeparerVille()
df['Ville'] = df.Nom_du_Fichier.apply(SeparerVille)
df.head()
```

```python
def SeparerDate(c):
    date = c.Nom_du_Fichier[len(c.Ville):c.Nom_du_Fichier.find(".txt")]
    return date
```

```python
df['Date'] = df.apply(SeparerDate, axis = 1)
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%b%d_%Y")
df.head()
```

```python
df.drop("Nom_du_Fichier", axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Année
df['Année'] = df.Date.dt.year
# Mois
df['Mois'] = df.Date.dt.month
# Jour (Jour du mois)
df['Jour_du_mois'] = df.Date.dt.day
# Jour de la semaine (0 pour lundi et 6 pour dimanche)
df['Jour_de_la_semaine'] = df.Date.dt.weekday

df.head()
```

```python
import numpy as np
```

```python
df["Weekend"] = np.where((df.Jour_de_la_semaine == 5) | (df.Jour_de_la_semaine == 6), "Oui", "Non")
df
```

```python
df.to_csv("cleaned-speeches.csv", index=False)
```
